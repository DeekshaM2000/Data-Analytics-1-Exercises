```python
#Importing the libraries
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#loading the Iris data set
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
print(df.head(30))
#Displaying the first 30 rows of the dataframe
sample = df.sample(n=30, random_state=42)
print(sample)

sepal_length_means = []
for i in range(100):
  sample = df.sample(n=30, random_state=i)
  mean_sepal_length = sample['sepal length (cm)'].mean()
  sepal_length_means.append(mean_sepal_length)
```

```python
#Histogram plotting
plt.hist(sepal_length_means, bins=10, edgecolor='black')
plt.xlabel('Mean Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of 100 Sample Means of Sepal Length')
plt.show()
```

```python
def systematic_sampling(df, step):
    indexes = np.arange(0, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample
```

```python
def systematic_sampling(df, percentage):

    step = int(len(df) / (len(df) * (percentage / 100)))
    return df.iloc[::step]
sampled_df = systematic_sampling(df, 20)
```

```python
# scatter plot of the original data
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], label='Original Data', alpha=0.5)

# catter plot of the sampled data
plt.scatter(sampled_df['sepal length (cm)'], sampled_df['sepal width (cm)'], label='Sampled Data', alpha=0.8, marker='o', edgecolors='black')

# Set plot labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Original vs. Sampled Data (Systematic Sampling)')

plt.legend()

# Displaying the plot
plt.show()
```
