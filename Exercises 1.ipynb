```python
#descriptive statistical analysis
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
file_path='/content/loan_data.csv'
data=pd.read_csv(file_path)
mean=np.mean(data['credit_score'])
print("Mean=",mean)
Median=np.median(data['credit_score'])
Mode = data['credit_score'].mode()[0]
print("Median=",Median)
print("Mode=",Mode)
Var=np.var(data['credit_score'])
print("Variance=",Var)
SD=np.std(data['credit_score'])
print("Standard Deviation=",SD)
range= np.ptp(data['credit_score'])
print("Range=",range)
```

```python
#Histogram plotting for Credit Score
plt.figure(figsize=(10,6))
plt.hist(data['credit_score'],bins=20,edgecolor='red',color='black')
plt.suptitle('Histograms of Credit Score',fontsize=18)
plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```
